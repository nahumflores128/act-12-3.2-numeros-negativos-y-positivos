```python
# para cuando es verdadero (si)
# identificar si es negativo o positivo
# datos de entrada Num,
# si es falso mostrar un signo negativo
# si es verdadero mostrar signo positivo
print("--- Determinación de Signo ---")
num = float(input("Ingrese el número (NÚM): "))
if num >= 0:
    R = "POSITIVO"
else:
    R = "NEGATIVO"
print("El número", num, "es:", R)
print("--- Proceso Finalizado ---")
```
